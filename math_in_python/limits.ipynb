```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
fx = x**3

lim_pnt = 1.5
lim = sym.limit(fx, x, lim_pnt)

print(lim)
display(Math('\\lim_{x\\to %g} %s = %g' %(lim_pnt, sym.latex(fx), lim)))
```

```python
fxx = sym.lambdify(x, fx)
xx = np.linspace(-5, 5, 200)

plt.plot(xx, fxx(xx))
plt.plot(lim_pnt, lim, 'ro')
plt.show()
```

```python
fx = (x**2) / (x - 2)
fxx = sym.lambdify(x, fx)

xx = np.linspace(1, 3, 102)

lim_pnt = 2
lim = sym.limit(fx, x, lim_pnt, dir='-')

plt.plot(xx, fxx(xx))
display(Math('\\lim_{x\\to %g^-} %s = %g' %(lim_pnt, sym.latex(fx), lim)))
```

```python
fx = (sym.E**(-x)) * (sym.sqrt(x + 1))
gx = sym.cos(x + sym.sin(x))

fxx = sym.lambdify(x, fx)
gxx = sym.lambdify(x, gx)

lim_pnt = 5

xx = np.linspace(-1, 5, 200)

lim = sym.limit(fx / gx, x, lim_pnt)
plt.plot(xx, fxx(xx))

display(Math(rf'\lim_{{x \to {lim_pnt}}} \frac{{{sym.latex(fx)}}}{{{sym.latex(gx)}}} = {float(lim): .4f}'))
```
