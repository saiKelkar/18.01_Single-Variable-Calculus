```python
def exponent(x, y):
    display(Math(f'{x}^{y} = {x**y}'))

def division(x, y):
    display(Math(f'\\frac{x}{y} = {x / y}'))

def main():
    i = int(input('input X: '))
    j = int(input('input Y: '))

    
    display(Math(f'\\text {{Press "1" to compute }} {i}^{j} \\text {{ or "2" to compute }} \\frac{i}{j}'))
    func_switch = input(' ')
    if func_switch == "1":
        print(exponent(i, j))
    elif func_switch == "2":
        print(division(i, j))
    else:
        print("Invalid selection!")
```

```python
main()
```

```python
import random
rand_int = random.randint(1, 100)
print(rand_int)

guess_num = int(input('Guess the number between 1 and 100: '))
while guess_num != rand_int:
    if guess_num > rand_int:
        print('Guess lower!')
    else:
        print('Guess higher!')
    guess_num = int(input('Guess again: '))
print(f'Got it! The right number was {rand_int} and your final guess was {guess_num}')
```
